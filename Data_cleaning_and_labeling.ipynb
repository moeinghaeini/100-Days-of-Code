| **Feature Group**      | **Feature Name**                                     | **Derived From**                       | **Why It's Important**                               |
| ---------------------- | ---------------------------------------------------- | -------------------------------------- | ---------------------------------------------------- |
| 🕓 Time                | `hour`, `part_of_day`, `is_weekend`                  | Timestamp                              | Screen usage varies by time (e.g., evening patterns) |
| 🔆 Light Sensor        | `ls_total`                                           | R + G + B + IR                         | Measures brightness, screens emit light              |

| 📏 Proximity           | `is_near` (proximity < threshold)                    | Proximity sensor                       | Detects if screen is within visible range            |
| 📐 Accelerometer       | `acc_magnitude`                                      | √(acc\_x² + acc\_y² + acc\_z²)         | Detects user motion/stability                        |
|                        | `tilt_angle`                                         | acc\_x, acc\_y, acc\_z                 | Reflects head or body posture toward screen          |
| 🔁 Gyroscope           | `gyro_magnitude`                                     | √(gyro\_x² + gyro\_y² + gyro\_z²)      | Detects rotation/motion while looking at screen      |
|                        | `rotation_variance`                                  | Rolling std of gyro\_x/y/z             | Measures stability of view                           |
| 🧾 Categorical Context | `device_type`                                        | "Device"                               | Different devices have different light signatures    |
|                        | `screen_type`                                        | "Screen type"                          | Phone vs. tablet vs. monitor                         |
|                        | `ambient_light_condition`                            | Pre-labeled / engineered               | Bright/dark environment affects screen visibility    |
|                        | `context`                                            | Label encoded (e.g., sitting, walking) | Helps interpret motion/light data                    |
| 🎯 Target              | `is_looking_at_screen`                               | Label                                  | Supervised learning target variable                  |


🕓 Time	hour, part_of_day, is_weekend	Timestamp	Screen usage varies by time (e.g., evening patterns)

```python
import pandas as pd
import numpy as np

# Example: assume 'df' is your DataFrame with a "Timestamp" column
def extract_time_features(df):
    # Convert to datetime if not already
    df['timestamp'] = pd.to_datetime(df['timestamp'])

    # Extract basic time features
    df['hour'] = df['timestamp'].dt.hour
    df['minute'] = df['timestamp'].dt.minute
    df['second'] = df['timestamp'].dt.second
    df['day_of_week'] = df['timestamp'].dt.dayofweek  # 0 = Monday, 6 = Sunday
    df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)

    # Part of day (custom buckets)
    def part_of_day(hour):
        if 5 <= hour < 12:
            return 'morning'
        elif 12 <= hour < 17:
            return 'afternoon'
        elif 17 <= hour < 21:
            return 'evening'
        else:
            return 'night'

    df['part_of_day'] = df['hour'].apply(part_of_day)

    # Optional: cyclical encoding (for time-based ML models)
    df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
    df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)

    return df
```

```python
# 🔆 Light Sensor Features
def extract_light_sensor_features(df):
    # Calculate total light sensor value
    df['ls_total'] = df['Light Sensor Red'] + df['Light Sensor Green'] + df['Light Sensor Blue'] + df['Light Sensor IR']

    # Calculate color ratios
    # Add a small epsilon to the denominator to avoid division by zero
    epsilon = 1e-9
    df['red_ratio'] = df['Light Sensor Red'] / (df['ls_total'] + epsilon)
    df['green_ratio'] = df['Light Sensor Green'] / (df['ls_total'] + epsilon)
    df['blue_ratio'] = df['Light Sensor Blue'] / (df['ls_total'] + epsilon)
    df['ir_ratio'] = df['Light Sensor IR'] / (df['ls_total'] + epsilon)

    return df

# Apply the function to your DataFrame
df3 = extract_light_sensor_features(df3)
df4 = extract_light_sensor_features(df4)
df5 = extract_light_sensor_features(df5)
df6 = extract_light_sensor_features(df6)

# Display the updated DataFrame with new features
df3.head()
```
